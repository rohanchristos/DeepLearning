```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Example data (independent variable x, dependent variable y)
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]).reshape(-1, 1)
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
```

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Data Scatter Plot')
plt.show()
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
plt.scatter(x, y, color='blue', label='Actual data')
plt.plot(x, y_pred, color='red', label='Regression line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

```python
r_sq = model.score(x, y)
print(f"Coefficient of determination (R^2): {r_sq}")
print(f"Slope (b_1): {model.coef_[0]}")
print(f"Intercept (b_0): {model.intercept_}")
```

```python
new_x = np.array([[2]])
pred_y = model.predict(new_x)
print("Predicted y for x=2:", pred_y)
```
